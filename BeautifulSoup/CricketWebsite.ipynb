# Importing Libraries

```python
from bs4 import BeautifulSoup
import requests
import lxml
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# Url of Website

```python
url = 'https://www.cricket.com/'
```

# Getting Request

```python
req = requests.get(url)
```

```python
data = req.text
```

```python
soup = BeautifulSoup(data , 'lxml')
```

```python
soup
```

```python
soup.prettify()
```

# Head Tag of the Website

```python
head = soup.find_all('head')
```

```python
for i in head:
    print(i.text)
```

# Title of Website

```python
title = soup.find_all('title')
```

```python
for i in title:
    print(i.text)
```

```python
for i in title:
    print(i.text)
```

# h2 tags of website

```python
h2 = soup.find_all('h2')
```

```python
h2
```

```python
for i in h2:
    print(i.text)
```

# Text of all span tags

```python
span = soup.find_all('span')
```

```python
for i in span:
    print(i.text)
```

# Text in div

```python
div = soup.find_all('div')
```

```python
for i in div:
    print(i.text)
```

# links in website

```python
links = soup.find_all('a')
```

```python
links
```

```python
for link in links:
    print(link.get('href'))
```

# Images in Website

```python
img = soup.find_all('img')
```

```python
img
```

```python
for image in img:
    print(image.get('src'))
```

```python
for i in img:
    link = 'https://cricket.com'+i.get('src')
    print(link)
```
